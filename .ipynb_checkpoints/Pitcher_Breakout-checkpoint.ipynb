# --- PITCHER BREAKOUT PREDICTION PROJECT ---

## Importing libraries

```python
import pandas as pd
import sklearn
import xgboost as xgb
import matplotlib.pyplot as plt 
import seaborn as sns
```

## Importing excel files 

```python
Pitching_2023 = pd.read_excel('Pitching_2023.xlsx', skiprows=1)
Pitching_2024 = pd.read_excel('Pitching_2024.xlsx', skiprows=1)

print(Pitching_2023.head())

print(Pitching_2024.head())
```

## Fixing Names 

```python
# For 2023
Pitching_2023[['Last', 'First']] = Pitching_2023['last_name, first_name'].str.split(', ', expand=True)
Pitching_2023['Name'] = Pitching_2023['First'] + ' ' + Pitching_2023['Last']
Pitching_2023 = Pitching_2023.drop(columns=['First', 'Last', 'last_name, first_name'])

# For 2024
Pitching_2024[['Last', 'First']] = Pitching_2024['last_name, first_name'].str.split(', ', expand=True)
Pitching_2024['Name'] = Pitching_2024['First'] + ' ' + Pitching_2024['Last']
Pitching_2024 = Pitching_2024.drop(columns=['First', 'Last', 'last_name, first_name'])

print(Pitching_2023.head())

print(Pitching_2024.head())
```

## Merging 2024 ERA with 2023 Data

```python
era_2024 = Pitching_2024[['Name', 'p_era']].rename(columns={'p_era': 'ERA_2024'})
print(era_2024.head())

Pitching_2023 = Pitching_2023.merge(era_2024, on='Name', how='left')

print(Pitching_2023.head())
```

## Filtering Pitchers by age and batters faced

```python
Pitching_2023_Filtered = Pitching_2023[
    (Pitching_2023['player_age'] <= 27) & 
    (Pitching_2023['pa'] >= 150) & 
    (Pitching_2023['pa'] <= 500)
]

Pitching_2023_Filtered = Pitching_2023_Filtered.dropna(subset=['ERA_2024'])


print(Pitching_2023_Filtered.head())
```

## Building the model 

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor

features = ['k_percent', 'bb_percent', 'wobadiff', 'hard_hit_percent', 'in_zone_percent', 'whiff_percent', 'groundballs_percent', 'flyballs_percent']  
X = Pitching_2023_Filtered[features]
y = Pitching_2023_Filtered['ERA_2024']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)

xgb_model = XGBRegressor(
    n_estimators=500,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=21
)

xgb_model.fit(X_train, y_train)

```

## Predicting 2025 ERA Using the Model

```python
Pitching_2024_Filtered = Pitching_2024[
    (Pitching_2024['player_age'] <= 27) & 
    (Pitching_2024['pa'] >= 150) & 
    (Pitching_2024['pa'] <= 500)
]

X_2024 = Pitching_2024_Filtered[features]

print(Pitching_2024_Filtered.head())

# Creating a copy so it leaves me alone 

Prediction_Data_2024 = Pitching_2024_Filtered.copy()
Prediction_Data_2024['Predicted_ERA_2025'] = xgb_model.predict(X_2024)
```

## Plotting the top 10 rookies

```python
Rookies_2024 = pd.read_excel('Rookies_2024.xlsx', skiprows=1)

rookies = Prediction_Data_2024[Prediction_Data_2024['Name'].isin(Rookies_2024['Name'])].copy()
print(rookies.head())

top15_rookies = rookies.sort_values('Predicted_ERA_2025').head(15)

plt.figure(figsize=(10,6))
sns.barplot(
    x='Predicted_ERA_2025', 
    y='Name', 
    data=top15_rookies,
    color='skyblue'  # single color instead of palette
)

plt.title('Top 15 Projected Rookies for 2025 (Predicted ERA)')
plt.xlabel('Predicted ERA 2025')
plt.ylabel('Name')
plt.tight_layout()
plt.show()
```

```python


non_rookies = Prediction_Data_2024[~Prediction_Data_2024['Name'].isin(Rookies_2024['Name'])].copy()

top15_non_rookies = non_rookies.sort_values('Predicted_ERA_2025').head(15)

plt.figure(figsize=(10,6))
sns.barplot(
    x='Predicted_ERA_2025', 
    y='Name', 
    data=top15_non_rookies,
    color='skyblue'  # single color instead of palette
)

plt.title('Top 15 non-rookie bounceback Pitchers for 2025 (Predicted ERA)')
plt.xlabel('Predicted ERA 2025')
plt.ylabel('Name')
plt.tight_layout()
plt.show()
```

## Model Importance 

```python
importances = xgb_model.feature_importances_


feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,  # or whatever your feature DataFrame is
    'Importance': importances
})


feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)


print(feature_importance_df)
```
